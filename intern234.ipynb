```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
hiv_data = pd.read_csv(r"C:\Users\ADMIN\Downloads\HIV data 2000-2023.csv")
```

```python
poverty_data = pd.read_excel(r"C:\Users\ADMIN\Downloads\multidimensional_poverty.xlsx")
```

```python
print(hiv_data.columns)
```

```python
hiv_data.columns = hiv_data.columns.str.strip()
```

```python
print(hiv_data.columns)
```

```python
hiv_data['Value'] = hiv_data['Value'].astype(str)
hiv_data['Value'] = hiv_data['Value'].str.replace(' ', '').str.extract(r'(\d+)')[0]
hiv_data['Value'] = pd.to_numeric(hiv_data['Value'], errors='coerce')
hiv_data['Value'].fillna(0)
hiv_data.dropna(subset=['Value'], inplace=True)
hiv_data['Value'] = hiv_data['Value'].astype(int)
print("Column 'Value' not found in dataset.")
```

```python
top_hiv_countries = hiv_data.groupby('Location')['Value'].sum().nlargest(int(len(hiv_data) * 0.75)).index.tolist()
```

```python
print(hiv_data.head())
print(poverty_data.head())
```

```python
top_hiv_data = hiv_data[hiv_data['Location'].isin(top_hiv_countries)]
```

```python
plt.figure(figsize=(16, 10))
sns.lineplot(data=top_hiv_data, x='Period', y='Value', hue='Location', marker='o')
plt.title('Trend of HIV Cases in Countries Countributing to 75% of Global Burden (2000-2023)')
plt.xlabel('Year')
plt.ylabel('Estimated Number of People Living with HIV')
plt.xticks(rotation=45)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16, 10))
sns.lineplot(data=top_hiv_data, x='Period', y='Value', hue='ParentLocationCode', marker='o')
plt.title('Trend of HIV Cases by WHO Region (2000-2023)')
plt.xlabel('Year')
plt.ylabel('Estimated Number of People Living with HIV')
plt.xticks(rotation=45)
plt.legend(title='WHO Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
