```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the mortality data
under_five_data = pd.read_csv(r"C:\Users\ADMIN/Downloads\under_five mortality rate.csv")
neonatal_data = pd.read_csv(r"C:\Users\ADMIN\Downloads\dataset_datascience.csv")

# List of EAC countries
eac_countries = ["Burundi", "Kenya", "Rwanda", "South Sudan", "Tanzania", "Uganda", "Democratic Republic of the Congo"]

# Filter data for EAC countries
under_five_eac = under_five_data[under_five_data['Geographic area'].isin(eac_countries)]
neonatal_eac = neonatal_data[neonatal_data['Geographic area'].isin(eac_countries)]

# Load shapefiles for each EAC country
kenya = gpd.read_file(r"C:\Users\ADMIN\Downloads\shapefile.shp\gadm41_KEN_0.shp")
uganda = gpd.read_file(r"C:\Users\ADMIN\Downloads\shapefile.shp\gadm41_UGA_1.shp")
tanzania = gpd.read_file(r"C:\Users\ADMIN\Downloads\shapefile.shp\gadm41_TZA_0.shp")
somalia = gpd.read_file(r"C:\Users\ADMIN\Downloads\shapefile.shp\gadm41_SOM_0.shp")
rwanda = gpd.read_file(r"C:\Users\ADMIN\Downloads\shapefile.shp\gadm41_RWA_0.shp")
burundi = gpd.read_file(r"C:\Users\ADMIN\Downloads\shapefile.shp\gadm41_BDI_0.shp")
south_sudan = gpd.read_file(r"C:\Users\ADMIN\Downloads\shapefile.shp\gadm41_SSD_0.shp")
democratic_republic_of_the_congo = gpd.read_file(r"C:\Users\ADMIN\Downloads\shapefile.shp\gadm41_COD_0.shp")

# Combine all shapefiles into a single GeoDataFrame
shapefiles = [kenya, uganda, tanzania, somalia, rwanda, burundi, south_sudan, democratic_republic_of_the_congo]
combined_shapefile = gpd.GeoDataFrame(pd.concat(shapefiles, ignore_index=True))

# Check the columns in the combined shapefile
print(combined_shapefile.columns)

# Merge the combined shapefile with under-five mortality data
merged_data_under_five = combined_shapefile.merge(under_five_eac, left_on="COUNTRY", right_on="Geographic area", how="left")

# Merge the combined shapefile with neonatal mortality data
merged_data_neonatal = combined_shapefile.merge(neonatal_eac, left_on="COUNTRY", right_on="Geographic area", how="left")

# Visualize Under-five Mortality Rates
plt.figure(figsize=(12, 8))
merged_data_under_five.plot(column='OBS_VALUE', cmap='viridis', legend=True, edgecolor='black')
plt.title("Under-five Mortality Rates in EAC Countries")
plt.axis('off')
plt.show()

# Visualize Neonatal Mortality Rates
plt.figure(figsize=(12, 8))
merged_data_neonatal.plot(column='Observation Value', cmap='viridis', legend=True, edgecolor='black')
plt.title("Neonatal Mortality Rates in EAC Countries")
plt.axis('off')
plt.show()

# Calculate average trends for Under-five mortality
average_trend_under_five = under_five_eac.groupby('Series Year')['OBS_VALUE'].mean().reset_index()

# Calculate average trends for Neonatal mortality
average_trend_neonatal = neonatal_eac.groupby('Series Year')['Observation Value'].mean().reset_index()

# Plot average trends for Under-five mortality
plt.figure(figsize=(10, 6))
sns.lineplot(data=average_trend_under_five, x='Series Year', y='OBS_VALUE', marker='o')
plt.title("Average Under-five Mortality Rate Over Time")
plt.xlabel("Year")
plt.ylabel("Average Mortality Rate")
plt.grid()
plt.show()

# Plot average trends for Neonatal mortality
plt.figure(figsize=(10, 6))
sns.lineplot(data=average_trend_neonatal, x='Series Year', y='Observation Value', marker='o')
plt.title("Average Neonatal Mortality Rate Over Time")
plt.xlabel("Year")
plt.ylabel("Average Mortality Rate")
plt.grid()
plt.show()

if not under_five_eac.empty and 'TIME_PERIOD' in under_five_eac.columns:
    highest_under_five = under_five_eac.loc[under_five_eac['TIME_PERIOD'].idxmax()]
else:
    print("Cannot find the highest under-five mortality rate due to empty DataFrame or missing column.")

# Identify the country with the highest neonatal mortality rate
highest_neonatal = neonatal_eac.loc[neonatal_eac['Observation Value'].idxmax()]

# Print results
print("Country with the highest Under-five mortality rate:")
print(highest_under_five[['Geographic area', 'TIME_PERIOD']])

print("\nCountry with the highest Neonatal mortality rate:")
print(highest_neonatal[['Geographic area', 'Reference Date']])
```
